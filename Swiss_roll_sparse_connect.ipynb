```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances
from sklearn.decomposition import PCA
import graphtools
from scipy.sparse.csgraph import connected_components
from datetime import datetime
import networkx as nx


def plot_graph_connectivity(data, L, std, sigma, title, color='red'):
    fig, ax = plt.subplots(figsize=(6, 6))
    fig.subplots_adjust(left=0.15, right=0.95, top=0.92, bottom=0.12)  # fixed margins
    
    pca = PCA(n_components=2)
    data_2d = pca.fit_transform(data)
    
    A = -L.toarray()
    np.fill_diagonal(A, 0)
    A = np.maximum(A, 0)
    G = nx.from_numpy_array(A)
    
    components = list(nx.connected_components(G))
    n_components = len(components)
    cmap = plt.cm.tab10
    node_colors = np.zeros(len(data))
    for i, comp in enumerate(components):
        for node in comp:
            node_colors[node] = i
    
    for u, v, w in G.edges(data='weight'):
        ax.plot([data_2d[u, 0], data_2d[v, 0]],
                [data_2d[u, 1], data_2d[v, 1]],
                color='grey', alpha=1.0,
                linewidth=0.5, zorder=1)
    
    scatter = ax.scatter(data_2d[:, 0], data_2d[:, 1],
                         c=node_colors, cmap=cmap,
                         zorder=5, vmin=0, vmax=9)
    
    if n_components > 1:
        plt.colorbar(scatter, ax=ax, label='Component index')

    ax.set_title(title, fontsize=16)
    ax.set_xlabel('PC1', fontsize=14)
    ax.set_ylabel('PC2', fontsize=14)
    ax.tick_params(axis='both', labelsize=14)

    now = datetime.now()
    date = now.strftime("%Y-%m-%d")
    time = now.strftime("%H-%M-%S")
    plt.savefig(f'/Users/x/Documents/connectivity_MLSP_11_june/AAA_Reconnected_swirl_{date}_{time}_std_{std}_sigma_{sigma}.png',
                dpi=500, transparent=True)
    plt.show()
    
    return G, components


def build_affinity_matrix(O, N, sigma=1.0):
    dist_NO = pairwise_distances(N, O, squared=True)
    dist_ON = dist_NO.T
    K_NO = np.exp(-dist_NO / (2 * sigma**2))
    K_ON = np.exp(-dist_ON / (2 * sigma**2))
    K_NN = K_NO @ K_ON
    return K_NN


def sort_swiss(data, label):
    distances = np.linalg.norm(data, axis=1)
    sorted_idx = np.argsort(distances)
    return data[sorted_idx], label[sorted_idx]


def reference_selection(data, n_clusters, random_state):
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
    kmeans.fit(data)
    return kmeans.cluster_centers_


def generate_gaussian(data, num_gauss_points, std=0.5):
    gaussian_points = []
    for point in data:
        points = np.random.normal(loc=point, scale=std, size=(num_gauss_points, len(point)))
        gaussian_points.append(points)
    return np.vstack(gaussian_points)


def swirl(n, start, end, size, y_scale):
    u1 = np.random.uniform(start * np.pi, end * np.pi, n)
    v1 = np.random.uniform(0, size, n)
    x1 = u1 * np.cos(u1)
    y1 = v1
    z1 = u1 * np.sin(u1)
    c = np.vstack((x1, y1 * y_scale, z1)).T
    return c


def graph_connectivity(O, N, K_NN, t):
    D = np.diag(np.sum(K_NN, axis=1))
    P_NN = np.linalg.inv(D) @ K_NN
    P_NN_powered = np.linalg.matrix_power(P_NN, t)
    N_graph_connectivity = P_NN_powered @ N
    all_points = np.vstack([O, N_graph_connectivity])
    return N_graph_connectivity, all_points

num_data_points = 100
noise = 0.2
random_state = 42
num_gauss_points = 500
std_list = [3.0]
sigma_list = [3.0]
num_points_list = [25]

c0 = swirl(100, 1.0, 3.0, 20, 0.1)
filler = swirl(1250, 3.0, 4.0, 20, 0.1)
filler = reference_selection(filler, 5, random_state)
c1 = swirl(50, 4.0, 4.5, 20, 0.1)

n_c0 = len(c0)
n_filler = len(filler)
n_c1 = len(c1)
 
original_data = np.vstack([c0, filler, c1])
colors = ['red'] * n_c0 + ['grey'] * n_filler + ['red'] * n_c1

pca = PCA(n_components=2)
pca_coords = pca.fit_transform(original_data)

fig, ax = plt.subplots(figsize=(6, 6))
ax.scatter(pca_coords[:, 0], pca_coords[:, 1], c=colors)
ax.set_xlabel('PC1', fontsize=14)
ax.set_ylabel('PC2', fontsize=14)
ax.set_title('d) Sparse Section Swiss roll', fontsize=16)
ax.tick_params(axis='both', labelsize=14)
now = datetime.now()
date = now.strftime("%Y-%m-%d")
time = now.strftime("%H-%M-%S")
#plt.savefig(f'/Users/x/Documents/connectivity_MLSP_11_june/AAA_START_Reconnected_swirl_{date}_{time}.png', dpi=500, transparent = True)
plt.show()

n_components = 10
band = 1.0
while n_components > 1:
    graph = graphtools.Graph(original_data, use_pygsp=True, bandwidth=band)
    graph.compute_laplacian("normalized")
    L = graph.L
    n_components, labels = connected_components(csgraph=L, directed=False, connection='weak')
    print(f"The graph is composed of {n_components} connected component(s). band = {band}")
    band = band + 0.1

final_band_original = band - 1
G, components = plot_graph_connectivity(
    original_data, L,
    std='none',
    sigma='none',
    title ='e) Without Graph Connectvity',
    color='red'
)


for std in std_list:
    for sigma in sigma_list:
        for num_gauss_points in num_points_list:
            generated_points = generate_gaussian(filler, num_gauss_points, std)
            affinity_matrix_NN = build_affinity_matrix(O=filler, N=generated_points, sigma=1.0)
            t_list = [1]
            for t in t_list:
                N_graph_connectivity, _ = graph_connectivity(filler, generated_points, affinity_matrix_NN, t)
                all_points = np.vstack([original_data, N_graph_connectivity])

                band = 1.0
                n_components = 10
                while n_components > 1:
                    graph = graphtools.Graph(all_points, use_pygsp=True, bandwidth=band)
                    graph.compute_laplacian("normalized")
                    L = graph.L
                    n_components, labels = connected_components(csgraph=L, directed=False, connection='weak')
                    print(f"The graph is composed of {n_components} connected component(s). band = {band}")
                    band = band + 0.1

                final_band_gc = band - 1

                final_band_original = band - 1
                G, components = plot_graph_connectivity(
                    all_points, L,
                    std=std,
                    sigma=sigma,
                    title ='f) With Graph Connectvity',
                    color='blue',
                )
```
